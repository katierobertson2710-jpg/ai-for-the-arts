# Part 4: Your Creative Design (Weeks 9-12) 

Large language models (LLM) are a type of artificial intelligence designed to understand and generate human-like text based on the data it has been trained on. The purpose of this notebook will be to present a design for an application that uses a large language model, as well as a critical refection of the social implications of this application. 

## Idea for Application 
A chatbot assistant that aids in helping skincare concerns. Users would input the issues they have with their skin e.g., dry skin, and the chatbot would return a response with information on types of skincare products that would help address their issues. This application will learn through reinforcement learning in which the model will be built to select the best action given an issue that leads to the most optimal outcome. This model would be trained on vast amounts of text data from dermatology sources including articles, medical textbook, as well as being trained on ingredient databases to detect what ingredients in skincare products would address specific issues. This would help the application to be able to learn patterns for what kinds of products address what kinds of skin issues. The application would be capable of answering questions on skin issues, prompting follow-up questions to find out more about the user’s issues to get as much information as possible, and generating recommendations and routines. 

### How this design will showcase the interaction between humans and AI
This design will showcase human-AI interaction through the input of a user prompt (in this case a skin concern), the application would be trained to respond with an explanation for the best ways to treat the concern based on the pervious data it has studied. The application would then respond with follow-up questions to discover the user’s pre-existing medical and skin conditions, budget preferences, etc., any kind of additional information that would help the application to refine the issue down to be able to suggest the best possible products. This iterative dialogue to ask for clarification loops back into the model and the application will continue to train on the information it has received. This is then a form of self-directed learning when the chatbot has to respond to users with dermatology concepts which goes back into the algorithm. 

This application has the potential to be inclusive of underrepresented groups as there is a serious lack of diverse representation in dermatological care. Historically, medical training has focused on how skin conditions have looked on lighter skin tones. This has had significant negative implications for people with black and brown skin as a lack of interest in the different ways skin issues present in people of colour can results in inaccurate assessments and delayed treatment (British Skin Foundation, 2025). Including a diverse range of skin tones in dermatology training is necessary for understanding how skin conditions present in different skin tones (Narla et al, 2023). AI training models present significant potential for expanding access to dermatological care. By training this model on diverse datasets, this application can be a tool for people whose skin issues have continually been dismissed by the medical community. This could have massive benefits for inclusive dermatology by reducing the risk of misdiagnosis in users with black and brown skin

### Critical reflection of social/cultural implications of proposed application 
My proposed application will be a chatbot that utilises LLM to discuss skin conditions with users in order to provide them with skincare product recommendations that suit their skin type. This will be beneficial to people with black and brown skin tones who are continually underrepresented to act as a dermatological resources to aid them. 

There is cause for concern as models that use existing databases to train could result in skewed or even inaccurate suggestions if these databases are not continually updated. This application will work around that by regularly auditing datasets and using bias detection tools such as fairness metrics and adverbial testing. Most importantly this application will aim to only train on data from diverse and representative datasets that include a wide range of demographics (Chapman University, 2025).

This application will aim to be transparent with its data usage. Passive collection of conversational data in AI models can reduce trust that users hold in the application which could make potential users hesitant to consult the chatbot as they may feel the descriptions of their skin issues are private and want to know exactly what happens to their personal data. This model will be completely transparent to its users for how it will store their conversational data so no users feel their privacy has been breached (Radanliev, 2025). 

It is important that this application is sensitive of the different ways skin conditions present in different skin tones, as well as the different ways various skincare products react to skin of all shades. By consulting medical professionals of all skin tones in the design process of this application, it would help to ensure that all users of all skin shades may benefit from the application. 

To conclude, this skin condition chatbot utilises large language model capabilities to enhance learning of dermatological conditions for the purpose of striving to be inclusive of all skin tones that require product recommendations.
